### Handle outliers

```python
sns.boxplot(df['Age']);
```

```python
for colname,colvals in df.iteritems():
    if df[colname].dtype!=object and colname!='Miles':
        Q1=df[colname].quantile(q=0.25)
        Q3=df[colname].quantile(q=0.75)
        IQR=Q3-Q1
        low=Q1-(1.5*IQR)
        up=Q3+(1.5*IQR)
        df.loc[(df[colname]<low) | (df[colname]>up) ,colname]=df[colname].median()
    else:
        continue
```

```python
sns.boxplot(df['Age'])
```

```python
#removed outliers
```

```python
df=pd.get_dummies(df,drop_first=True)
df.head()
```

### standardization

```python
#select x and y
x=df[['Fitness']]
y=df['Miles']
df.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
std=scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(std,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
mlr=LinearRegression()
```

```python
mlr.fit(x_train,y_train)
```

### Prediction

```python
print('Intercept:',mlr.intercept_)
print('Coefficient:',mlr.coef_)
```

```python
y_train_pred=mlr.predict(x_train)
y_test_pred=mlr.predict(x_test)
```

### Performance

```python
from sklearn import metrics
import numpy as np
print('Mean absolute error:',metrics.mean_absolute_error(y_test,y_test_pred))
print('Mean squared error:',metrics.mean_squared_error(y_test,y_test_pred))
print('Root mean squared error:',np.sqrt(metrics.mean_squared_error(y_test,y_test_pred)))
print('R-2:',metrics.r2_score(y_test,y_test_pred))
```

```python
# print the R-squared value for the model
from sklearn.metrics import accuracy_score
print('R squared value of the model: {:.2f}'.format(mlr.score(std,y)*100))
```

**Conclusion:** 60.81% of model fit in regression analysis
